```python
import os
import pandas as pd
```

## Extract

```python
time_series_covid19 = pd.read_csv('time_series_covid19_confirmed_global.csv')

time_series_covid19
```

## Transform

```python
time_series_covid19=time_series_covid19.melt(id_vars=['Province/State', 'Country/Region','Lat','Long'], 
        var_name='date', 
        value_name='number')
```

```python
time_series_covid19['date'] = pd.to_datetime(time_series_covid19['date'], format='%m/%d/%y').dt.strftime('%Y-%m-%d')
```

```python
time_series_covid19
```

## Load

```python
from sqlalchemy import create_engine
source = create_engine('mysql+mysqlconnector://test:test123@db/test')
```

```python
source.execute("DROP TABLE IF EXISTS time_series_covid19_confirmed_global")
print("Table dropped... ")
```

```python
time_series_covid19.to_sql('time_series_covid19_confirmed_global', con=source, schema='test', if_exists='append',index=False,chunksize=10000)
```
